## Question 3:

What combination of borough and year hard the largest number of homicide victims?

```python
con.sql("""
    SELECT
        boroname,
        year,
        SUM(CAST(num_victim AS INTEGER)) AS total_victims
    FROM nyc_homicides
    GROUP BY boroname, year
    ORDER BY total_victims DESC
    LIMIT 1;
""").show()
```

## Question 4:

Which subway station has the largest number of shooting victims within a 300 meter radius?


```python
con.sql("""
    SELECT
        s.name AS station_name,
        s.borough,
        SUM(CAST(h.num_victim AS INTEGER)) AS total_shooting_victims
    FROM nyc_subway_stations AS s
    JOIN nyc_homicides AS h
        ON ST_Distance(s.geom, h.geom) <= 300
    GROUP BY s.name, s.borough
    ORDER BY total_shooting_victims DESC
    LIMIT 1;
""").show()
```

## Question 5

what is the most densely populated residential street in new york? (total population of intersecting census blocks divided by the total length of the street)

```python
con.sql("""
    WITH street_stats AS (
        SELECT
            s.name,
            SUM(b.popn_total) AS total_pop,
            ST_Length(s.geom) AS street_length
        FROM nyc_streets AS s
        JOIN nyc_census_blocks AS b
            ON ST_Intersects(s.geom, b.geom)
        WHERE s.type = 'residential'
        GROUP BY s.name, s.geom
    )

    SELECT
        name,
        total_pop,
        street_length,
        (total_pop / street_length) AS density
    FROM street_stats
    WHERE street_length > 0
    ORDER BY density DESC
    LIMIT 5;
""").show()
```

## Question 6
How many tiles are contained within new york city neighbourhoods?

```python
con.sql("""
    SELECT count(DISTINCT o.quadkey) AS tile_count
    FROM ookla_nyc AS o
    JOIN nyc_neighborhoods AS n
        ON ST_Intersects(ST_GeomFromText(o.geometry), n.geom);
""").show()
```

## Question 7

Which neighbourhood has the highest average download speed? use 'within' rather than intersects, and ignore tiles with fewer than 3 tests.

```python
con.sql("""
    SELECT
        n.name AS neighborhood_name,
        AVG(o.avg_d_kbps) / 1000 AS avg_download_speed_mbps
    FROM ookla_nyc AS o
    JOIN nyc_neighborhoods AS n
        ON ST_Within(ST_GeomFromText(o.geometry), n.geom)
    WHERE o.tests >= 3
    GROUP BY n.name
    ORDER BY avg_download_speed_mbps DESC
    LIMIT 1;
""").show()
```

## Question 8



What is the name of the subway station closest to the tile with the highest download speed?

```python
con.sql("""
    WITH fastest_tile AS (
        SELECT ST_Centroid(ST_GeomFromText(geometry)) AS tile_geom
        FROM ookla_nyc
        ORDER BY avg_d_kbps DESC
        LIMIT 1
    )

    SELECT
        s.name AS station_name,
        ST_Distance(s.geom, f.tile_geom) AS distance
    FROM nyc_subway_stations AS s, fastest_tile AS f
    ORDER BY distance ASC
    LIMIT 1;
""").show()
```

## Question 9

Which neighbourhood has the highest number of devices per capita? Use only spatial intersections.

```python
con.sql("""
    WITH neighborhood_devices AS (

        SELECT
            n.name,
            n.geom,
            SUM(o.devices) AS total_devices
        FROM nyc_neighborhoods AS n
        JOIN ookla_nyc AS o
            ON ST_Intersects(n.geom, ST_GeomFromText(o.geometry))
        GROUP BY n.name, n.geom
    ),
    neighborhood_pop AS (

        SELECT
            n.name,
            SUM(b.popn_total) AS total_population
        FROM nyc_neighborhoods AS n
        JOIN nyc_census_blocks AS b
            ON ST_Intersects(n.geom, b.geom)
        GROUP BY n.name
    )
    SELECT
        d.name AS neighborhood,
        d.total_devices,
        p.total_population,
        (CAST(d.total_devices AS FLOAT) / NULLIF(p.total_population, 0)) AS devices_per_capita
    FROM neighborhood_devices d
    JOIN neighborhood_pop p ON d.name = p.name
    WHERE p.total_population > 0
    ORDER BY devices_per_capita DESC
    LIMIT 1;
""").show()
```

## Question 10


What is the average upload speed for tiles that are within 200 meters of L train stops and intersect with census blocks where the black population is above 50%?

```python
con.sql("""
    WITH l_train_stops AS (
        SELECT geom
        FROM nyc_subway_stations
        WHERE routes LIKE '%L%'
    ),

    black_majority_blocks AS (
        SELECT geom
        FROM nyc_census_blocks
        WHERE (CAST(popn_black AS FLOAT) / NULLIF(popn_total, 0)) > 0.5
    )

    SELECT
        AVG(o.avg_u_kbps) / 1000 AS avg_upload_speed_mbps
    FROM ookla_nyc AS o
    JOIN l_train_stops AS s
        ON ST_Distance(ST_GeomFromText(o.geometry), s.geom) <= 200
    JOIN black_majority_blocks AS b
        ON ST_Intersects(ST_GeomFromText(o.geometry), b.geom);
""").show()
```
